```python
import os
import boto3
import ipywidgets as widgets
from IPython.display import display, clear_output

# Set environment credentials tiruan untuk keamanan lokal LocalStack
os.environ["AWS_ACCESS_KEY_ID"] = "test"
os.environ["AWS_SECRET_ACCESS_KEY"] = "test"
os.environ["AWS_DEFAULT_REGION"] = "us-east-1"

# 1. KONFIGURASI ALAMAT LOCALSTACK (DENGAN KREDENSIAL DUMMY)
ENDPOINT_URL = "http://localhost:4566"
REGION = "us-east-1"

s3_client = boto3.client(
    's3', 
    endpoint_url=ENDPOINT_URL, 
    region_name=REGION, 
    aws_access_key_id='test', 
    aws_secret_access_key='test'
)
ec2_client = boto3.client(
    'ec2', 
    endpoint_url=ENDPOINT_URL, 
    region_name=REGION, 
    aws_access_key_id='test', 
    aws_secret_access_key='test'
)

output_area = widgets.Output()

# Widget S3 & EC2
txt_s3_name = widgets.Text(placeholder='bucket-name', description='S3 Name:')
btn_create_s3 = widgets.Button(description='Create Bucket', button_style='success')
btn_delete_s3 = widgets.Button(description='Delete Bucket', button_style='danger')

lbl_ec2 = widgets.HTML("<h3><b>[ SAIFUL AMRI - EC2 MANAGER ]</b></h3>")
btn_create_ec2 = widgets.Button(description='Luncurkan EC2', button_style='primary')
txt_ec2_id = widgets.Text(placeholder='i-xxxxxx', description='EC2 ID:')
btn_delete_ec2 = widgets.Button(description='Delete EC2', button_style='warning')

# ==========================================
# FUNGSI MANAJEMEN STORAGE S3 (KREATIVITAS SAIFUL)
# ==========================================
def buat_bucket_saiful(b):
    with output_area:
        clear_output()
        nama_bucket = txt_s3_name.value.strip()
        if not nama_bucket:
            print("❌ Nama bucket tidak boleh kosong!")
            return
        try:
            s3_client.create_bucket(Bucket=nama_bucket)
            print(f"✅ Sukses membuat S3 Bucket baru: {nama_bucket}")
        except Exception as e:
            print(f"❌ Gagal membuat bucket: {str(e)}")

def hapus_bucket_saiful(b):
    with output_area:
        clear_output()
        nama_bucket = txt_s3_name.value.strip()
        if not nama_bucket:
            print("❌ Nama bucket tidak boleh kosong!")
            return
        try:
            s3_client.delete_bucket(Bucket=nama_bucket)
            print(f"🗑️ Sukses menghapus S3 Bucket: {nama_bucket}")
        except Exception as e:
            print(f"❌ Gagal menghapus bucket: {str(e)}")

btn_create_s3.on_click(buat_bucket_saiful)
btn_delete_s3.on_click(hapus_bucket_saiful)

# ==========================================
# FUNGSI MANAJEMEN INSTANCE EC2 (KREATIVITAS SAIFUL)
# ==========================================
def luncurkan_ec2_saiful(b):
    with output_area:
        clear_output()
        try:
            response = ec2_client.run_instances(ImageId='ami-7e62731f', MinCount=1, MaxCount=1, InstanceType='t2.micro')
            instance_id = response['Instances'][0]['InstanceId']
            txt_ec2_id.value = instance_id
            print(f"🚀 Sukses meluncurkan EC2 Instance: {instance_id}")
        except Exception as e:
            print(f"❌ Gagal meluncurkan EC2: {str(e)}")

def hancurkan_ec2_saiful(b):
    with output_area:
        clear_output()
        instance_id = txt_ec2_id.value.strip()
        if not instance_id:
            print("❌ EC2 ID tidak boleh kosong!")
            return
        try:
            ec2_client.terminate_instances(InstanceIds=[instance_id])
            print(f"🔥 Sukses menghancurkan EC2 Instance: {instance_id}")
        except Exception as e:
            print(f"❌ Gagal menghapus EC2: {str(e)}")

btn_create_ec2.on_click(luncurkan_ec2_saiful)
btn_delete_ec2.on_click(hancurkan_ec2_saiful)

# Penyusunan Layout Tampilan Vertikal & Horizontal Box
lbl_s3 = widgets.HTML("<h3><b>[ SAIFUL AMRI - S3 MANAGER ]</b></h3>")
box_s3 = widgets.VBox([lbl_s3, txt_s3_name, widgets.HBox([btn_create_s3, btn_delete_s3])])
box_ec2 = widgets.VBox([lbl_ec2, btn_create_ec2, widgets.HBox([txt_ec2_id, btn_delete_ec2])])

# Menampilkan Identitas Mahasiswa dan UI Utama berbentuk Tab
print("Aplikasi Manajemen Cloud - SAIFUL AMRI (NIM: 32602500074)")
display(widgets.Tab(children=[box_s3, box_ec2], titles=('S3 Storage', 'EC2 Compute')))
display(output_area)
```
