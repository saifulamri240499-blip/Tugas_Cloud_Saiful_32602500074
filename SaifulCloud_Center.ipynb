```python
# 0. Set ENV & Persiapan Library
import datetime as dt
import ipywidgets as widgets
from IPython.display import display, clear_output

print("Library berhasil dimuat.")
```

```python
# 1. Konfigurasi Alamat & Koneksi
MODE = "demo"  # Pilihan: "demo" atau "ministack"
ENDPOINT = "http://localhost:4566"
REGION = "us-east-1"

DUMMY_CREDS = {
    "aws_access_key_id": "local-dummy-key",
    "aws_secret_access_key": "local-dummy-secret"
}

print(f"Mode aktif : {MODE}")
print(f"Region     : {REGION}")
```

```python
# 2. Widget S3 & EC2
btn_create_bucket = widgets.Button(description="Buat S3 Bucket", button_style="success")
btn_list_bucket = widgets.Button(description="Cek S3 Bucket", button_style="info")
btn_launch_ec2 = widgets.Button(description="Jalankan EC2", button_style="warning")

txt_ec2_id = widgets.Text(value="i-0abcd12345ef", description="EC2 ID:")
btn_delete_ec2 = widgets.Button(description="Delete EC2", button_style="danger")

output_area = widgets.Output()

print("Widget S3 & EC2 berhasil diinisialisasi.")
```

```python
# 3. Fungsi Manajemen Storage & Compute
def handle_create_bucket(b):
    with output_area:
        clear_output()
        print("[SIMULASI S3] Berhasil membuat bucket cloud baru.")

def handle_list_bucket(b):
    with output_area:
        clear_output()
        print("[SIMULASI S3] Daftar bucket: ['bucket-tugas-saiful-32602500074']")

def handle_launch_ec2(b):
    with output_area:
        clear_output()
        print("[SIMULASI EC2] Instance server berhasil dijalankan (ID: i-0abcd12345ef)")

def handle_delete_ec2(b):
    with output_area:
        clear_output()
        print("[SIMULASI EC2] Sukses menghancurkan EC2 Instance")

# Menghubungkan tombol dengan fungsi backend
btn_create_bucket.on_click(handle_create_bucket)
btn_list_bucket.on_click(handle_list_bucket)
btn_launch_ec2.on_click(handle_launch_ec2)
btn_delete_ec2.on_click(handle_delete_ec2)

print("Fungsi backend berhasil dipasang.")
```

```python
# 4. Penyusunan Layout Panel Kendali (Tab Menu)
print("Aplikasi Manajemen Cloud - SAIFUL AMRI (NIM: 32602500074)")

box_s3 = widgets.VBox([
    widgets.HTML("<b>[ SAIFUL AMRI - S3 MANAGER ]</b>"),
    widgets.HBox([btn_create_bucket, btn_list_bucket])
])

box_ec2 = widgets.VBox([
    widgets.HTML("<b>[ SAIFUL AMRI - EC2 MANAGER ]</b>"),
    widgets.HBox([btn_launch_ec2, txt_ec2_id, btn_delete_ec2])
])

tab_ui = widgets.Tab(children=[box_s3, box_ec2])
tab_ui.set_title(0, 'S3 Storage')
tab_ui.set_title(1, 'EC2 Compute')

display(tab_ui)
display(output_area)
```
